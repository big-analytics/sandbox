```python
m = folium.Map(
    location=[50.907657, 4.373416],
    zoom_start=14
)

for idx, row in g_df.iterrows():
    infoLabel=str(row["G_LOCATION"]) + "\n" + str(row["BEDROOMS"]) + " bedrooms\n" + str(row["LIVING_AREA"]) + " living area\n" + str(row["LAND"]) + " land\n" + str(row["PRICE"])
    #print infoLabel
    
    folium.Marker(
        location=[row.loc["G_LAT"], row.loc["G_LNG"]],
        popup=infoLabel.decode("utf8"),
        icon=folium.Icon(color='beige', icon='fa-home', prefix='fa')
    ).add_to(m)

m
```
